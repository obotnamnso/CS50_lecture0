### Reading in the data from VenueDailyVisitor.csv 


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('https://tinyurl.com/ChrisCoDV/001183787/VenueDailyVisitors.csv', index_col=0)
data.index = pd.to_datetime(data.index)
```

### Perform some basic statistical analysis using .head(), .tail(), and .describe()

```python
print(data)
print(data.head())
print(data.tail())
print(data.describe())
print(data.sum())
# print(data.shape)
```

1. There are 40 Venues lebelled 'YXF', 'PXI', 'WRL' ---- 'XPE', ''VLS', 'QRY'.
2. The duration of the dataset[Venues]  is a year (365) from 2019-01-10 to 2019-12-31.
3. The description for Venue 'YXF'
  * Mean = 75.71
  * Min = 38.00
  * Max = 122.00



# Sort the data according to the maximum value of each row, smallest first

```python
data = data.reindex(data.max(axis=1).sort_values().index)
print(data.max(axis=1))
# print(data.head())
```

1. The lowest Venue visitded was on 2019-04-19 which was 377
2. The highest Venue visited was on 2019-07-02 which was 647 

# Sort the data according to the sum of each column, largest first


```python
data = data.reindex(data.sum().sort_values(ascending=False).index, axis=1)
print(data.sum())
print(data.head())
```

1. 'RDA' 181,771 has the highest numbers of visit
2. 'YVW' 4,966 has the lowest numbers of visit 

## Sort the data according to the Average of each column, largest first


```python
data = data.reindex(data.mean().sort_values(ascending=False).index, axis=1)
print(data.mean())
print(data.head())
```

## Adding a column which shows the average venue visits for the entire venue. 

```python
data['Average'] = data.mean(axis=1)
print(data.head())

selected = ['RDA', 'SPF', 'Average']
data = data[selected]

data.plot.line()
plt.show()
```

```python
data = data.loc[pd.to_datetime('2019-02-01'): pd.to_datetime('2019-02-28')]
print(data.head())

data.plot.line()
# plt.show()

data.plot.box()

data.plot.density()
```

```python
print(data.head())
print(data.sum())

plt.figure(figsize=(8, 8))
x_pos = np.arange(len(data.columns))
plt.bar(x_pos, data.sum(), align='center')
plt.xticks(x_pos, data.columns)
plt.xlabel('Venue', fontsize=18)
plt.ylabel('Visitors', fontsize=18)
plt.title('High Volume Venues', fontsize=20)
plt.show()
```
